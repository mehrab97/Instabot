## Question 5 Like/Unlike posts
### 1. Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
### 2. Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

### 1. Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

```python
#Locating the search box and typing 'dilsefoodie' in the searchbox

box=driver.find_element_by_xpath('//div[contains(@class,"LWmhU")]/input')
box.clear()
box.send_keys('dilsefoodie')
```

```python
#Locating dilsefoodie in the instagram handles displayed and clicking on it
time.sleep(2)
dil=driver.find_element_by_xpath('//div[contains(@class,"fuqBx")]/a[1]')
dil.click()
```

```python
#Locating the 1st post and clicking on it
time.sleep(2)
post1=driver.find_element_by_class_name('_9AhH0')
post1.click()
```

```python
#To keep count of the number of posts liked
count=0

#Locating the like button and liking the first post
time.sleep(2)
like_btn=driver.find_element_by_xpath('//span[contains(@class,"fr66n")]/button')
like_btn.click()
count+=1
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
#Repeating the same process for 29 other posts
for i in range(29):
    #Moving onto the next post by clicking on the next button
    wait=WebDriverWait(driver,30)
    lap=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'coreSpriteRightPaginationArrow')))
    lap.click()
    
    #Locating the like button and clicking on it
    wait=WebDriverWait(driver,30)
    lap1=wait.until(EC.presence_of_element_located((By.XPATH,'//span[contains(@class,"fr66n")]/button')))
    lap1.click()
    count+=1
    
#Check if liked the posts or not
if count==30:
    print('I have liked all the posts')
    
```

### 2. Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
#We're currently on the 30th post.

#To keep count of the number of posts unliked
count_=0

#Locating the like button and clicking on it again so as to unlike the post
time.sleep(2)
like_btn=driver.find_element_by_xpath('//span[contains(@class,"fr66n")]/button')
like_btn.click()
count_+=1
```

```python
#Repeating the same process for 29 other posts
for i in range(29):
    #Moving onto the previous post by clicking on the previous button
    wait=WebDriverWait(driver,30)
    lap=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'coreSpriteLeftPaginationArrow')))
    lap.click()
    
    #Locating the like button and clicking on it so as to unlike the post
    wait=WebDriverWait(driver,30)
    lap1=wait.until(EC.presence_of_element_located((By.XPATH,'//span[contains(@class,"fr66n")]/button')))
    lap1.click()
    count_+=1
    
#After executing this, we reach the first post which has been unliked using the above for loop as well

#Check if unliked the posts or not
if count_==30:
    print('Unliked all the posts')
```

```python
#To find details of the profile pic section which will also show the story if there is one
story=driver.find_element_by_xpath('//div[contains(@class,"RR-M-")]/canvas')

try:
    #If there is a story and I haven't yet seen it
    if story.get_attribute('height')=='182':
        print('I am yet to view the story')
    
    #If there is a story and I have seen it
    elif story.get_attribute('height')=='174':
        print('I have already seen the story')
    
#If there is no story
except NoSuchElementException:
    print('The user has no story')
```
